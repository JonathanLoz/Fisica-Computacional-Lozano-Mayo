```julia
f(x)=x^3-x^2-x-1
df(x)=3x^2-2x-1
function NEWTON(f,df,xo,tolerancia)
Xk=[]
ERROR=[]
    while abs(f(xo))>tolerancia
        xo=xo-f(xo)/df(xo)
        push!(Xk,xo)
    end
    ERROR=(Xk-xo)/xo,Xk,xo
end     
```

```julia
#Método de la bisección

function signo(x)

    if x>0
        return 1
    end
    if x<0
        return -1
    end
    if x==0
        return 0
    end
end
        
function raizbiseccion(f,a,b,tolerancia)
Aprox=[]    
c=(b+a)/2 
     if signo(f(a))==signo(f(b))
        print("Es posible que no haya raíz en este intervalo")
        end
     if f(c)==0 
     return f(c)
     end
   
    while abs(f(c))>tolerancia
        if signo(f(a))==signo(f(c))
            a=c
            b=b
        end
        if signo(f(a))!=signo(f(c))
            a=a
            b=c
        end
        c=(b+a)/2
        push!(Aprox,c)
    end
    error=abs.((Aprox-c)/c)
    return error,Aprox,c
end 
```

```julia
errorb1,aprox1,raiz1=raizbiseccion(f,0,10,1e-8)
errorb2,aprox2,raiz2=raizbiseccion(f,-10,10,1e-8)
errorb3,aprox3,raiz3=raizbiseccion(f,1,2,1e-8)
errorb4,aprox4,raiz4=raizbiseccion(f,1,45,1e-8)
```

Usar la función NEWTON.(F,df,intervalo,tolerancia) ya cumple lo requerido por el ejercicio 2 pero para facilitar la lectura de las entradas, definiré la funcion newtonraiz que podra ser evaluada en un intervalo pero son los regresará la raiz para cada condicion en ese intervalo

```julia
function newtonraiz(f,df,xo,tolerancia)
Xk=[]
ERROR=[]
    while abs(f(xo))>tolerancia
        xo=xo-f(xo)/df(xo)
        push!(Xk,xo)
    end
    xo
end    
```

```julia
newtonraiz.(f,df,linspace(2,40,40),1e-5)
```

```julia
#Este es el metodo de newton con tolerancia \epsilon y que solo da 200 pasos
function newton200(f,df,xo,tolerancia)
Xk=[]
ERROR=[]
i=0    
    while abs(f(xo))>tolerancia && i<200
        i=i+1       
        xo=xo-f(xo)/df(xo)
        push!(Xk,xo) #aqui estamos guardando cada valor en un arreglo       
    end
    xo
end    
```

```julia
newton200.(f,df,linspace(2,20,10),1e-4)
```

```julia
using SymPy
x=symbols("x")
function NewtonSympy(f,xo) #Esta funcion admite una condicion inicial o un intervalo de donde toma las condiciones iniciales
Xk=[]
ERROR=[]
df=lambdify(diff(f(x),x))
    for i in 1:200
        xo=xo-f(xo)/df(xo)
        push!(Xk,xo)
    end    
    xo
end
```

```julia
NewtonSympy.(f,linspace(45,100,10)) #ejemplo con un intervalo que nos arrojará todos los pasos hasta que converja el metodo con cada condicion inicial
```
