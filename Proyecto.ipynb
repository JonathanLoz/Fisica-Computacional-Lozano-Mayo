__Problema__: Encontrar la masa límite de una enana blanca antes de convertirse en una estrella de neutrones o un agujero negro, este límite en la masa es conocido como límite de Chandrasekhar. 

__ Teoría__

En teoría cinética podemos encontrar que el número de partículas por unidad en el espacio fase está dada por (densidad numérica):


\begin{equation}
\frac{dN}{d^3x d^3p}=\frac{g}{\hbar^3}f\tag{1}
\end{equation}

donde $h^3$ es el volumen de la celda en el espacio fase,g es el número de estados de una particula con un determinado momento $p$ y $f$ es el número de ocupación promedio de una celda en el espacio fase.

Además la densidad de partículas en el espacio de momentos está dada por 

\begin{equation}
n=\int\frac{dN}{d^3x d^3p}d^3p\tag{2}
\end{equation}

Por otra parte la presión de un sistema con una distribución de momentos isotrópica está dada por 

\begin{equation}
P=\frac{1}{3}\int pv\frac{dN}{d^3x d^3p}d^3p\tag{3}
\end{equation}

Para un gas ideal en equilibrio sabemos que $f$ tiene una forma simple:

\begin{equation}
    f(E)=\frac{1}{e^{(E-\mu)/kT}\pm 1}\tag{4}
\end{equation}


Para un gas de fermiones completamente degenerado, es decir, con   ($T\rightarrow 0, \frac{\mu}{kT}\rightarrow \infty$) $f(E)$ tiende a las siguientes igualdades:

\begin{equation}
f(E)=
\begin{cases} 
      1 & E\leq E_F \\
      0 & E > E_F
      \end{cases}\tag{5}
\end{equation}

para estos gases $\mu$ es llamada la energía de Fermi $E_F$.

__Ecuación de estado de un gas ideal de Fermi completamente degenerado.__

En última instancia una enana blanca o una estrella de neutrones aisladas tienden a enfriarse, es decir, tienden a la temperatura cero, y es la presión asociada a la materia con $T=0$ la que evita que estas estrellas sufran un colapso gravitacional. Es por esta razón que estudiar la ecuación de estado de un gas ideal de Fermi completamente degenerado nos permite tener un modelo sencillo para las estrellas de neutrones y enanas blancas. 

Supongamos que tenemos un gas de una sola especia, digamos electrones para los cuales $g=2S+1=2$, de este modo, si definimos el momento de fermi $p_F$ como 

\begin{equation}
 E_F=(p^2_Fc^2+m_ec^4)^\frac{1}{2}\tag{6}
 \end{equation}
 
 Tendríamos que $n_e$ está dada por la expresión $(2)$, y sustituyendo $(1)$ se obtiene:
 
 
\begin{equation}
n_e=\frac{2}{\hbar^3}\int_0^{p_F}4\pi p^2dp=\frac{8\pi}{3\hbar^3}p_F^3\tag{7}
\end{equation}


Luego definiendo el momento de Fermi adimensional $x_F$ como

\begin{equation}
 x_F=\frac{p_F}{m_ec}\tag{8}
 \end{equation}

podemos expresar $n_e$ en términos del momento de fermi adimensional obteniendo 


\begin{equation}
 n_e=\frac{1}{3\pi^2\lambda^3_e}x^3\tag{9}
 \end{equation}

con $\lambda_e=\frac{\hbar}{m_ec}$ la longitud de onda de Compton del electrón.

Así mismo la presión del gas está dada por la siguiente integral

\begin{align}
P_e=&\frac{1}{3}\frac{2}{\hbar^3}\int_0^{p_F}\frac{p^2c^2}{(p^2_Fc^2+m_ec^4)^\frac{1}{2}}4\pi p^2 dp \\=&\frac{8\pi m^4_ec^5}{3\hbar^3}\int_0^x\frac{x^4}{(1+x^2)^\frac{1}{2}}dx\\=& \frac{m_ec^2}{\lambda^3_e}\phi(x)\tag{10}
\end{align}


donde 

\begin{equation}
\phi(x)=\frac{1}{8\pi^2}[x(1+x^2)^\frac{1}{2}(\frac{2x^2}{3}-1)+ln[x+(1+x^2)^\frac{1}{2}]]\tag{11}
\end{equation}

en el régimen ultra relativista, es decir cuando $\rho >> 6*10^{15} g cm^{-3}$, 
 $\phi(x) \rightarrow \frac{x^5}{15\pi^2}$ y la presión puede ser escrita en su forma politrópica. 
 
 \begin{equation}
 P=K\rho^\Gamma\tag{12}
 \end{equation}
 
 con $\Gamma=\frac{4}{3}$  y  $K=1.2293*10^{15} cgs$



__ Ecuación de equilibrio hidrostático__

Si idealizamos la distribución de una estrella de tal manera que tenga simetría esférica entonces la masa de la estrella dentro de un radio r está dada por:

\begin{equation}
m(r)=\int_0^r \rho 4\pi r^2 dr\tag{13}
\end{equation}

o en forma diferencial

\begin{equation}
\frac{dm(r)}{dr}=4\pi r^2 \rho\tag{14}
\end{equation}


Por otra parte si consideramos la atracción gravitacional que siente una masa $dm$ debido a $m(r)$ y la presión neta que se ejerce sobre $dm$, es decir, $-P(r+dr)+P(r)$, luego si tomamos un elemento diferencial de area del fluido $dA$ perpendicular a la dirección radial, tendrémos que en equilibro la fuerza gravitacional y la fuerza debida a la presión que se ejerce deben igualarse y tenemos entonces:

\begin{align}
\frac{Gm(r)}{r^2}dm=&-[P(r+dr)-P(r)]dA\\
=&\frac{dP}{dr}dA\\
\end{align}

Y despejando se obtiene la ecuación de equilibrio hidrostático para la estrella

\begin{equation}
\frac{dP}{dr}=-\frac{Gm(r)\rho}{r^2}\tag{15}
\end{equation}


Veamos que en general sabemos la presión como función de la densidad, pero esta realación involucra un logaritmo así que no podemos despejar esta ecuación para obtener la densidad, así que las ecuaciones $(14)$ y $(15)$ deben resolverse de manera numérica, esto es, a partir de una densidad central podemos obtener la presión central y con esta usar la ecuación diferencial y obtener un paso de la ecuación de la masa también usando algún método de integración, sin embargo para obtener el siguiente paso cuando resolvemos las ecuaciones simultaneamente debemos tambier obtener la siguiente densidad pues esta depende del radio, eso lo podemos lograr implementando algún método de inversión de ecuaciónes de manera aproximada como lo es el método de Newton. 

En lo que sigue implementaremos esta solución computacional, para encontrar el límite de Chandrasekhar, que nos indica la masa con la cual una enana blanca colapsa pues la presión electrónica ya no es capaz de sustentar el empuje gravitacional y la estrella se compacta.


###### Solución computacional

```julia
using Plots
theme(:dark)
include("herramientas.jl") #Algunas herramientas programadas previamente.
#------------------------------------------------------------------------#
#método de Newton implementado para resolver el estado de la enana blanca.
function NEWTON(f,y,xo,h)
    n=0
    while abs(f(xo)-y)>h
        n=n+1
        df=(f(xo+h)-f(xo))/h
        xo=xo-(f(xo)-y)/df
        if n>100
            break
        end
    end
    xo
end    
```

__ Constantes __

```julia
MasaSol=1.98855e33 #masa solar
A=12.0 #peso atómico
z=6.0 #número atómico
μ=A/z #peso molecular promedio
UMA=1.6605389e-24 
me=9.11e-28   #masa electron
c=3e10        #velocidad de la luz
h=6.626e-27
hb=h/(2.*pi) #constante de Planck reducida
lc=hb/(me*c) #Longitud de onda de Compton
G=6.67259e-8 #Constante de gravitación universal
rhoC = 3e6   #Densidad central para la estrella
ne = rhoC/me #Densidad de electrones(electrones por unidad de volumen)

#Consideraciones para la ecuación de estado en forma politropica. 
#Caso Ultra relativista 
gur=4.0/3.0
Kur=1.2435e15/μ
ρUR(press)=(press/Kur)^(1/gur)
#Caso no relativista 
PT(y)=Kur*y^(gur)
gnr=5.0/3.0
Knr=1.036e13

```

```julia
#Funciones auxiliares

ϕ(x)=(1/(8.*pi^2))*(x*((1+x^2)^(.5))*((2.*x^(2.)/3.)-1.)+log(x+(1+x^2.)^.5))
P(x)=((me*c^2.)/(lc^3.))*ϕ(x) #Presion de la estrella en términos de la xF el momento adimensional de fermi
pF(x)=(abs(3.*pi^2.*(hb^3)*x))^(1./3.) #momento de Fermi
xF(x)=x/(me*c) #momento adimensional de Fermi
Chandrasekhar=ones(4170).*1.43145 #limite de Chandrasekhar
ρNR(press)=μ*(press/Knr)^(1/gnr) #rho aproximada como politropo
neF(x)=(x^3./(3.*pi^2.*hb^3.)) #n en terminos de xF
rhoF(x)=μ*UMA*x #rho en términos del momento adimensional de Fermi
```

__Esta función me da la masa ($m$), la presión ($P$) y la densidad ($\rho$) limite de una estrella
dada la densidad central y la presión central, así como la masa en $r=0$__.

```julia
function Estrella(i,n)
#Valores iniciales
#------------------------------------------------------------------------------#
rhoC = n*10^i #Densidad central para la estrella
ne = rhoC/me #Densidad de electrones(electrones por unidad de volumen)
press = P(xF(pF(ne))) #Presión central de la estrella
n = 0              #contador del método de runge kutta
dr = 1*10^5   #Paso del método de Runge Kutta
r = 0       #radio del método de Runge kutta
m = 0        #masa del método de runge kutta
x = xF(pF(ne)) #Momento correspondiente a la densidad central
#---------------------------------------------------------------------------
M=[] #Arreglo de Masa
R=[] #Arreglo de Radio
Pe=[] #Arreglo de Presión
Rho=[] #Arreglo de densidad
    
#Proceso iterativo del método de Runge Kutta de orden 4    
while press>0
 
 r=r+dr
 n=n+1   
 rho=rhoF(neF(me*c*NEWTON(P,press,x,0.01)))       
 push!(Pe,press)
 push!(Rho,rho)       
     
 
 rho1=rhoF(neF(me*c*NEWTON(P,press,x,0.01))) #Push 
 k1m=(4*pi*(r^2)*rho1)
 M1=m+(k1m*dr*0.5)
 k1p=-(G*m*rho1)/(r^2.)
 P1=press+(k1p*dr*0.5)
 
 
 rho2=rhoF(neF(me*c*NEWTON(P,P1,x,0.01)))
 k2m=(4*pi*((r+dr*0.5)^2)*rho2)
 M2=m+(k2m*dr*0.5)
 k2p=-(G*M1*rho2)/((r+dr*0.5)^2.)
 P2=press+(k2p*dr*0.5)
 
 
 
 
 rho3=rhoF(neF(me*c*NEWTON(P,P2,x,0.01)))   
 k3m=(4*pi*((r+dr*0.5)^2)*rho3)
 M3=m+(k3m*dr*0.5)
 k3p=-(G*(M2)*rho3)/((r+dr*0.5)^2.)
 P3=press+(k3p*dr*0.5)
 
    
 rho4=rhoF(neF(me*c*NEWTON(P,P3,x,0.01)))
 k4m=(4*pi*((r+dr)^2)*rho4)
 k4p=-(G*M3*rho4)/((r+dr)^2.)
 



 
 m=m+(dr/6.)*(k1m+2.*(k2m+k3m)+k4m)
 press=press+(dr/6.)*(k1p+2.*(k2p+k3p)+k4p)
  
    
 push!(M,m/MasaSol)
 push!(R,r/1e5)  
 
    end
R,M,Pe,Rho
end
y = xF(pF(ne));
```

```julia
#Condiciones centrales
rho=3*10^6
ne = rho/me
presion=P(xF(pF(ne)))
y = xF(pF(ne))
xo=[0.1,presion]

#Sistema de ecuaciones para usarse en un método de Runge Kutta de Cuarto orden previamente creado-----#
f(x,t)=[4*pi*(t^2)*rhoF(neF(me*c*NEWTON(P,x[2],y,0.01))),
    -G*x[1]*rhoF(neF(me*c*NEWTON(P,x[2],y,0.01)))/(t^2)]  
#------------------------------------------------------------------------#
# Resolvemos la ecuación con Runge Kutta 4to orden
s3=herramientas.RungeKutta4th(f,xo,0.1,3e8,1e5)

#-------------------------------------------------------------------------#
# Iterador que crea las listas de Masa,Radio y Presión.
function Listas(s)
listM=[]
listP=[]
listR=[]
for i in 1:length(s[2])
    m=s[2][i][1]
    p=s[2][i][2]
    r=s[1][i]
    if isnan(p)==false || isnan(m)==false
        push!(listM,m/MasaSol)
        push!(listP,p)
        push!(listR,r/1e5)
    end
end
    return listM,listP,listR
end
#--------------------------------#
M3=Listas(s3)[1]
P3=Listas(s3)[2]
R3=Listas(s3)[3];
```

TRASFORMADA DE FOURIER  (en forma matricial)

```julia
function DFT(x)
    
    N = length(x)
    
    M = [exp(-2*im*pi*i*j/N) for i in 0:N-1, j in 0:N-1]
    
    x_T = M*x
   
    return x_T
end
```

```julia
function FFT(x)
    
    N=length(x)
   
    
    if N<=4
        
        return DFT(x)
        
    end
        
    X_impares = FFT(x[2:2:N])
    X_pares = FFT(x[1:2:N])

    C = [exp(-2*im*pi*k/N) for k in 0:N-1]
    
    XK = vcat(X_pares + [C[k]*X_impares[k] for k in 1:length(X_impares)], X_pares + [C[k+length(X_pares)]*X_impares[k] for k in 1:length(X_impares)])
    
    #me da los primeros k de 0 a longitud impares y la siguiente entrada me da los k que me faltan
    
    return XK
end
```

```julia

function IFFT(x_tilde)
    
    N=length(x_tilde)
    
    x_tilde_imag = imag(x_tilde)
    x_tilde_real = real(x_tilde)
    
    x_real = FFT(x_tilde_real)
    x_imag = FFT(-x_tilde_imag)

    YY = (x_real + im.*x_imag)/N
 
    return YY
end
```

```julia
using Plots
# Parámetros del intervalo en el espacio x
N1 = 2^12
h1 = 1/ N1

# Arreglo en el espacio x
x1 = collect(linspace(h1, 1, N1))

# Arreglo en el espacio k

k1 = [0:N1/2; -N1/2+1:-1]

# Función rho

function ρ(x)
    if abs(x - 0.5) < 0.05
        return 10
    else
        return 0
    end
end

ρv = ρ.(x1)

# Resolvemos en el espacio de Fourier
ϕt = -FFT(ρv) ./ k1.^2
ϕt[1] = 0      #Arreglamos en el caso k=0

# Aplicamos la transformada inversa
phi = real((IFFT(ϕt)))

#Graficamos
plot(x1, phi,label="ϕ(x)")
scatter!([0.5], [0], label = ("Carga en x=1/2"))
```
