Diagrama de bifircaciones y exponente de Lyapunov del mapeo $$ x_{n+1}=rx_n(1- x_n)$$

```julia
function mapeo1(xo,r,n) #estas rutinas definen los puntos del diagrama de bifurcaciones 
        for i in 1:n
        xo=r*xo*(1-xo)
        end
    xo
end
q1=linspace(0,1,1000)
m1=linspace(1,4,1000)
m2=linspace(1,3.9,1000)
H=[]
M1=[]
for i in 1:length(m1)
    push!(H,mapeo1.(q1,m1[i],1000))
    push!(M1,ones(length(q1))*m1[i])
end

function Lyapunov(xo,n,r)  # aquí hice el calculo del coeficiente de Lyapunov iterando
l=0    
for i in 1:n
        
    xo=r*xo*(1-xo)   
    l=l+log(abs(r*(1-2*xo)))
      
    end
    l/n
end
```

```julia
scatter(M1,H,label="",mc=:green,markersize=0.5,xlabel="r",ylabel="x",title="x",xticks=1:.5:4)
```

```julia
scatter(m1,Lyapunov.(0.5,1000,m2),markersize=0.5,xlabel="r",ylabel="λ",label="",xticks=1:.5:4)
```

Diagrama de bifircaciones y exponente de Lyapunov del mapeo $$ x_{n+1}=1-\mu x_n^2$$

```julia
function mapeo(xo,r,n)  #estas rutinas definen los puntos del diagrama de bifurcaciones 
        for i in 1:n
        xo=1-r*xo^2
        end
    xo
end
q=linspace(-1,1,1000)
m=linspace(0,2,1000)
m3=linspace(0.1,2,1000)
M=[]
G=[]
for i in 1:length(m)
    push!(G,mapeo.(q,m[i],1000))
    push!(M,ones(length(q))*m[i])
end

function Lyapunov2(xo,n,r)   # aquí hice el calculo del coeficiente de Lyapunov iterando
l=0    
for i in 1:n
        
    xo=1-r*xo^2   
    l=l+log(abs(-2*r*xo))
      
    end
    l/n
end
```

```julia
scatter(M,G,label="",markersize=0.005,xlabel="μ",ylabel="x",markershape=:circle,xticks=0:.5:2)   
```

```julia
scatter(m3,Lyapunov2.(0,1000,m3),markersize=0.5,xlabel="μ",ylabel="λ",label="",xticks=0:.5:2)
```
